```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.navy.mil.kr/mbshome/mbs/navy/subview.do?id=navy_010105000000'
res = requests.get(url)

try:
    res.raise_for_status()
except Exception as e:
    print(e)

soup = BeautifulSoup(res.text, "html.parser")

#with open("solder.csv", "w", encoding="utf8") as f:
#    f.write(soup.prettify())

solders_note = soup.find("ul", class_ = 'nv_imgList2')
solders = solders_note.find_all("li", class_ = 'nv_imgListLi2')
result = []
for solder in solders:
    result.append(solder.find("img")["alt"])

result = [i for i in reversed(result)] 

df = pd.DataFrame(result, columns=["역대 해군 대장들!!"])
df.index.name = "순위"
df.index = df.index + 1
df.to_csv("beach_solders.csv", encoding='utf-8-sig', mode='w', index=True)
df.to_excel("beach_solders2.xlsx")
```
