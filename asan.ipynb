### 분야랑 사람 이름 가져오기

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

profession_wrap = dict()

for i in range(100): 
    url = 'https://www.amc.seoul.kr/asan/staff/base/staffBaseInfoList.do?drEmpId=&deptTabIndex=&searchHpCd=D{:0>3}'.format(i + 1)
    res = requests.get(url)
    #print(i)
    try:
        res.raise_for_status() # getcode == 200
    except Exception as e:
        print(e)

    try:
        soup = BeautifulSoup(res.text, "html.parser")

        #with open("asan.csv", "w", encoding="utf8") as f:
        #    f.write(soup.prettify())
        tag_ul = soup.find("ul", class_ = 'serchlist_boxwrap')
        tag_li = tag_ul.find_all("li")
        wrap = tag_li[0].find("table", class_ = 'professionally_info').find("td").get_text(strip=True).split(',')[0]
        print(wrap)
        doctors = []
        for i in tag_li:
            doctor = i.find("p", class_ = "doctor_name").find("a").get_text(strip=True)
            print(doctor)
            doctors.append(doctor)
        profession_wrap[wrap] = doctors
    except:
        continue

    df = pd.DataFrame({ k: pd.Series(v) for k, v in profession_wrap.items()})
    df.head()
```
