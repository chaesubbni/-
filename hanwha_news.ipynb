### 한화 에어로스페이스 뉴스 개수 뽑기

```python
from selenium import webdriver
import time
```

```python
url = 'https://www.hanwhaaerospace.com/kor/media/newsroom/list.do'
wd = webdriver.Chrome()
wd.get(url)

while True:
    try:
        wd.find_element(webdriver.common.by.By.XPATH, '/html/body/div[3]/main/div/section[3]/div[3]/button/span[1]').click()
        time.sleep(1)
    except:
        pass
```

```python
from bs4 import BeautifulSoup

year = dict()

soup = BeautifulSoup(wd.page_source, "html.parser")

tag_lis = soup.find("ul", class_ = 'news-list').find_all("li")

for tag_li in tag_lis:
    y = tag_li.select_one("span.date").get_text()[:4]
    if y in year:
        year[y] += 1
    else:
        year[y] = 1
```
